```python
import pickle
import pandas as pd
```

```python
my_list = pd.read_pickle('./data/real_time_data2.pkl')
```

```python
len(my_list)
```

```python
final_df = pd.DataFrame(columns=["symbol", "day", "HrMin", "Second", "timestamp", "price"])
for i in range(len(my_list)):
    temp_df = my_list[i]
    final_df = pd.concat([temp_df, final_df])

final_df.drop_duplicates()
final_df
```

```python
META = final_df[final_df['symbol']=='META']
META
```

```python
META = META.drop(['Second', 'DateTime', 'timestamp'], axis=1)
```
